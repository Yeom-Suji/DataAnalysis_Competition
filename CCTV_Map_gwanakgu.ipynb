```python
from geopy.distance import geodesic
import folium

existing_markers = []  # 기존 마커 리스트

# 기존 데이터로부터 마커 생성
for i in range(len(data)):
    lat = data.loc[i, '위도']
    long = data.loc[i, '경도']
    name = data.loc[i, '안심 주소']
    
    # 기존 마커를 생성하고 리스트에 추가
    marker = folium.Marker(location=[lat, long], popup=name)
    existing_markers.append(marker)

def markers_within_radius(latitude, longitude, existing_markers):
    # 새로운 좌표에 빨간색 마커 추가
    new_marker_map = folium.Map(location=[latitude, longitude], zoom_start=17)
    folium.Marker(location=[latitude, longitude], popup='New Marker', icon=folium.Icon(color='red')).add_to(new_marker_map)

    # 기존 마커들과의 거리 계산
    markers_in_radius = []
    for marker in existing_markers:
        marker_location = marker.location
        coords_1 = (latitude, longitude)
        coords_2 = (marker_location[0], marker_location[1])
        
        # 좌표 간 거리 계산 (단위: 미터)
        distance = geodesic(coords_1, coords_2).meters
        
        # 반경 500m 내에 있는 경우 해당 마커들을 저장
        if distance <= 500:
            markers_in_radius.append(marker)

    # 기존 마커와 새로운 마커 모두 추가한 지도 반환
    for marker in markers_in_radius:
        marker.add_to(new_marker_map)
    
    return new_marker_map, len(markers_in_radius)

# 사용자로부터 좌표 입력 받기
user_latitude = float(input("위도를 입력하세요: "))
user_longitude = float(input("경도를 입력하세요: "))

# 함수 호출하여 새로운 마커 추가 및 반경 내 기존 마커 수 확인
new_marker_map, count = markers_within_radius(user_latitude, user_longitude, existing_markers)

# 결과 출력
print(f"500m 반경 내에 {count}개의 안심 CCTV가 있습니다.")
new_marker_map
```
